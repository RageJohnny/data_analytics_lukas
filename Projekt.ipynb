Alle benötigten Bibliotheken werden geladen.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Parquet file wird geladen und die ersten paar Zeilen angezeigt.
Link für Syntax: https://www.nyc.gov/assets/tlc/downloads/pdf/data_dictionary_trip_records_yellow.pdf

```python
data01 = pd.read_parquet('yellow_tripdata_2022-12.parquet')
data01.head()
```

```python
# get information about the loaded data
data01.info()
```

Bereinigen der Daten 

```python
# Entfernen von Duplikaten
data02 = data01.drop_duplicates()
```

```python
#Spalte store_and_fwd_flag entfernen
data03 = data02.drop('store_and_fwd_flag', axis=1,)
```

```python
# Entfernen von NaN
data04 = data03.dropna()
```

```python
# Überprüfen ob VendorId nur 1 oder 2 enthält rest löschen
data05 = data04[data04['VendorID'].isin([1, 2])]
```

```python
# Überprüfen ob tpep_pickup_datetime nur 2022-12 enthält rest löschen

# Stellen Sie sicher, dass 'tpep_pickup_datetime' und 'tpep_dropoff_datetime' im datetime Format ist
data05['tpep_pickup_datetime'] = pd.to_datetime(data05['tpep_pickup_datetime'])


# Definieren Sie den Start- und Endzeitpunkt
start_date = '2022-12-01 00:00:00'
end_date = '2022-12-31 23:59:59'

# Filtern Sie die Daten
data06 = data05[data05['tpep_pickup_datetime'].between(start_date, end_date)]
data06
```

```python
# Überprüfen ob tpep_pickup_datetime nur 2022-12 enthält rest löschen
# Definieren Sie den Start- und Endzeitpunkt

data05['tpep_dropoff_datetime'] = pd.to_datetime(data05['tpep_dropoff_datetime'])

start_date = '2022-12-01 00:00:00'
end_date = '2023-01-01 23:59:59'

# Filtern Sie die Daten
data07 = data06[data06['tpep_dropoff_datetime'].between(start_date, end_date)]

# Überprüfen Sie das Ergebnis
data07.info()
```

```python
# Filtern Sie die Daten, so dass 'tpep_pickup_datetime' immer vor 'tpep_dropoff_datetime' liegt
data08 = data07[data07['tpep_pickup_datetime'] < data07['tpep_dropoff_datetime']]

# Überprüfen Sie das Ergebnis
data08.info()
```

Pasenger Count (Spalte 4) wird erst für die Berechnung des Tips per Person bereinigt

```python
# Stellen Sie sicher, dass 'trip_distance' ein Float ist
data09['trip_distance'] = data08['trip_distance'].astype(float)

# Filtern Sie die Daten, so dass 'trip_distance' größer als 0 und kleiner als 500 ist
data10 = data09[(data09['trip_distance'] > 0) & (data09['trip_distance'] < 500)]

# Überprüfen Sie das Ergebnis
data10.info()
```

```python
# Finden Sie den Index des maximalen Werts in 'trip_distance'
max_value_index = data10['trip_distance'].idxmax()

# Zeigen Sie die Zeile mit dem maximalen Wert in 'trip_distance' an
max_value_row = data10.loc[max_value_index]

print("Die Zeile mit dem maximalen Wert in 'trip_distance' ist:")
print(max_value_row)
```

```python
# Berechnen Sie die Dauer zwischen 'tpep_pickup_datetime' und 'tpep_dropoff_datetime'
data10['Duration'] = data10['tpep_dropoff_datetime'] - data10['tpep_pickup_datetime']

# Definieren Sie die Reihenfolge der Spalten
cols = data10.columns.tolist()
cols.insert(cols.index('tpep_dropoff_datetime') + 1, cols.pop(cols.index('Duration')))

# Ordnen Sie die Spalten neu an
data10 = data10[cols]

# Überprüfen Sie das Ergebnis
data10.info()
```

```python
# Filter the data so that 'Duration' is greater than or equal to 1 minute and less than or equal to 3 hours
data11 = data10[(data10['Duration'] >= pd.Timedelta(minutes=1)) & (data10['Duration'] <= pd.Timedelta(hours=10))]

# Drop the rows where 'Duration' is greater than 3 hours or less than 1 minute
data12 = data11.drop(data11[(data11['Duration'] > pd.Timedelta(hours=3)) | (data11['Duration'] < pd.Timedelta(minutes=1))].index)

# Check the result
data12.info()
```

```python
# Calculate average speed in distance per hour
data12['Average_Speed'] = data12['trip_distance'] / (data12['Duration'].dt.total_seconds() / 3600)

# Reorder columns
cols = data12.columns.tolist()
cols.insert(cols.index('Duration') + 1, cols.pop(cols.index('Average_Speed')))
data13 = data12[cols]

# Check the result
data13.info()
```

```python
# Filtern Sie die Daten, so dass 'Average_Speed' größer als 1 und kleiner als 100 ist
data14 = data13[(data13['Average_Speed'] > 3.00) & (data13['Average_Speed'] < 100.00)]

# Überprüfen Sie das Ergebnis
data14.info()

```

```python

# Show the shortest duration
print("The ride with the shortest duration is:")
print(data14.loc[data14['Duration'].idxmax()])
```
